```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('temperature2.csv')
print(df.head())
```

```python
df["date1"] = pd.to_datetime(df["date"],dayfirst=True)
monthly = df.groupby(df["date1"].dt.to_period("M"))["meantemp"].mean()
plt.plot(monthly.index.astype(str),monthly.values,marker="o")
plt.xlabel("Months")
plt.ylabel("Temperature")
plt.title("Monthly Temperature")
plt.show()
```

```python
df["date1"] = pd.to_datetime(df["date"],dayfirst=True)
monthly = df.groupby(df["date1"].dt.to_period("M"))["rainfall"].mean()
plt.scatter(monthly.index.astype(str),monthly.values,marker="o")
plt.xlabel("Months")
plt.ylabel("Temperature")
plt.title("Monthly Temperature")
plt.show()
```
