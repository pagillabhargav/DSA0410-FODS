```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('onlinesales.csv')
print(df.head())
```

```python
df["Date"]=pd.to_datetime(df["InvoiceDate"],dayfirst=True)
df["Month_sales"]=df["Date"].dt.to_period("M")
print(df.head())
```

```python
df_2023 = df[df["Month_sales"].dt.year == 2023]
monthly=df_2023.groupby("Month_sales")["UnitPrice"].sum()
plt.bar(monthly.index.astype(str),monthly.values)
plt.xlabel("Months",)
plt.ylabel("Average Sales")
plt.title("Monthly Sales")
plt.xticks(rotation=90)
plt.show()
```

```python
jan_2023 = df[
    (df["Date"].dt.year == 2023) &
    (df["Date"].dt.month == 1)
]
daily_sales = jan_2023.groupby("Date")["UnitPrice"].sum()
plt.plot(daily_sales.index, daily_sales.values, marker="o")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.title("Daily Sales â€“ January 2023")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
